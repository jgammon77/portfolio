```python
"""
Josh Gammon
CS-6678 XTIA-24T2 Assignment 7
December 8th, 2024

This program contains two functions that are used - removePunctuation and the main function.
Main reads in the text then calls removePunctuation which returns the text without any special characters or punctuations.
The main function then removes non-English words using NLTK and the corpus module. The results are stored in a str.
To get the total unique words, I decided to first split the string into a list, then convert the list to a set.
A set contains only unique values so we can simply get the length of the set to find the total unique words.
I also created a pos_tagger function that takes in the sample text, removes punctuation and non-english words
then applies the tagger on the remaining words.

For output, I had the text with punctuations removed shown which is a lot of text.
Underneath all of that text there are the results after the non-English words are removed, stored in a list, and set.
Finally, I showed the length of the list and set so we can see the difference between the total words and unique words.
These numbers are buried underneath the text so please scroll all the way down for those results.

Final Note: NLTK probably does not have the most up to date English word list and eliminated some words that may
be considered like Sonnets and even names like William Shakespeare.
"""

import nltk
from nltk.corpus import words
from nltk.tokenize import word_tokenize
from collections import Counter
nltk.download('words')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
#nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
#stop_words = set(stopwords.words('english'))

####THIS FUNCTION REMOVES ALL PUNCTUATIONS AND SPECIAL CHARACTERS THAT MAY BE IN THE TEXT####
def removePunctuation(text):
  punctuations = '''!()-[]{};:'"\,<>./?^&$%*_'''

  for char in text.lower():
    if char in punctuations:
      text = text.replace(char, "")
  #print(text)
  return text

def main():
  with open("Shakespeare.txt", "r") as file:                      ### OPEN AND READ THE FILE
    text = file.read()
  input = removePunctuation(text)                                 ### TEXT IS PASSED INTO THE removePunctuation FUNCTION, TEXT WITH REMOVED PUNCTUATION IS THEN RETURNED
  words = set(nltk.corpus.words.words())                          ### NLTK PACKAGE COMES WITH BUILT-IN ENGLISH WORDS IN THE CORPUS MODULE, IT IS USED HERE TO REMOVE
  text2 = " ".join(w for w in nltk.wordpunct_tokenize(input) \
           if w.lower() in words or not w.isalpha())              ### NON-ENGLISH WORDS FROM THE TEXT.
  print(text2)

  print("---------------------------------OUPUT START-----------------------------------------------------------------------------------------------------------------------------")
  print("")
  print("-------------------------1. HERE IS THE LIST-----------------------------------------------------------------------------------------------------------------------------")
  print("")
  list = text2.split()                      ### OUTPUT FROM NON-ENGLISH TEXT REMOVAL RESULTS IN A MASS OF TEXT IN A STRING OBJECT, THIS IS SPLIT INTO A LIST
  list = [word.lower() for word in list]    ### MAKE ALL THE WORDS LOWERCASE
  print(list)
  print("")
  print((f'1. Number of words in the text (non-English words removed): {len(list)}'))  ### LENGTH OF THE LIST
  print("")

  print("-------------------2. HERE ARE THE UNIQUE SET OF WORDS SORTED ALPHABETICALLY------------------------------------------------------------------------------------------")
  uniqueSet = sorted(set(list))   ### LIST IS MOVED TO A SET. A SET ONLY CONTAINS UNIQUE VALUES SO THIS HOLDS ALL THE UNIQUE WORDS IN THE TEXT
  print(f'2B. {uniqueSet}')
  print("")
  print(f'2A. Number of unique words in the text (non-English words removed): {len(uniqueSet)}') ### LENGTH OF THE SET TELLS US THE NUMBER OF UNIQUE WORDS
  print("")

  #Count = Counter(list)
  #mostFrequent20 = Count.most_common(20)
  #print("")
  #print(f'---------------------------------------Top 20 words -------------------------------------------') ### TOP 20 WORDS
  #print(mostFrequent20)


def pos_tagger():

  testText = "Cupid laid by his brand and fell asleep. A maid of Dians this advantage found. And his lovekindling fire did quickly steep"
  testTextNoPunc = removePunctuation(testText)                                                  #### REMOVE PUNCTUATIONS FROM TEST TEXT
  words = set(nltk.corpus.words.words())
  testTextNoPuncNonEngRemoved = " ".join(w for w in nltk.wordpunct_tokenize(testTextNoPunc) \
           if w.lower() in words or not w.isalpha())                                            ### REMOVING NON-ENGLISH WORDS
  tokenize_text = sent_tokenize(testTextNoPuncNonEngRemoved)
  for word in tokenize_text:
    wordList = nltk.word_tokenize(word)
    taggedList = nltk.pos_tag(wordList)                                                         ### POS TAGGER APPLIED

    print("-------------------------------------------3. TAGGING APPLIED TO TEXT----------------------------------------------------------------------------------------------------")
    print("")
    print(f'HERE IS THE WORDLIST WITH NON-ENGLISH WORDS AND PUNCTUATIONS REMOVED: ')
    print(wordList)
    print("")
    print(f'HERE IS THE TAGGED LIST WITH NON-ENGLISH WORDS AND PUNCTUATIONS REMOVED: ')
    print(taggedList)

main()
pos_tagger()
```
