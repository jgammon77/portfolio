QUESTION #1 -> Calculate the death rates of all counties in AL, and sort the counties by the death rate in descending order.

```python
df = pd.read_csv('COVID-19 in Alabama-2.csv')
```

After reading in the dataset I like to look at the data in its current form. We can see that it created row 0 that does not contain any information, so this row will be removed. I will also remove the final row that contains a general count of the variables.

The dataset contains 4 variables (Counties, Cases, Total Tested By County, and Deaths) with 67 instances.

```python
df
```

```python
df = df.drop([0])
df = df.drop([68])
```

```python
df
```

```python
df.isnull().sum()
```

Now we see that there are no missing values within the dataset.

Calculate the death rates of all counties in AL, and sort the counties by the death rate in descending order

```python
##CONVERTING DATA TYPES FROM OBJECTS TO INTEGERS/STRING TYPES
df['Deaths'] = df['Deaths'].astype(int)
df['Cases'] = df['Cases'].astype(int)
df['Total Tested By County'] = df['Total Tested By County'].astype(int)
df['Counties'] = df['Counties'].astype(str)
```

```python
df.dtypes
```

```python
###In this barplot we can visualize the amount of deaths per county.

plt.figure(figsize=(15,5), dpi=200)
sns.barplot(data=df, x='Counties', y='Deaths',color='red')
plt.xticks(rotation=90)
plt.show()
```

```python
###In this barplot we can visualize the amount of cases per county.

plt.figure(figsize=(15,5), dpi=200)
sns.barplot(data=df, x='Counties', y='Cases',color='purple')
plt.xticks(rotation=90)
plt.show()
```

```python
###In this barplot we can visualize the amount of cases per county.

plt.figure(figsize=(15,5), dpi=200)
sns.barplot(data=df, x='Counties', y='Total Tested By County',color='green')
plt.xticks(rotation=90)
plt.show()
```

```python
###From these plots we can see that Jefferson County had the highest amount of deaths, cases, and testing.
###From this we discover a correlation between testing, positive diagnosees then deaths as directly proportaional
```

```python
###To test this theory, we can calculate the death rate for each county then compare.
#Death Rate = (Deaths/Cases)*100

df['Death Rate'] = (df['Deaths']/df['Cases'])*100
#df.sort_values(by='Death Rate', ascending=False)
```

```python
df
```

```python
###In this barplot we can visualize the Death Rate per county.

plt.figure(figsize=(15,8), dpi=200)
sns.barplot(data=df, x='Counties', y='Death Rate',color='orange', order=df.sort_values('Death Rate', ascending=False).Counties)
plt.xticks(rotation=90)
plt.show()
```

```python
df.sort_values('Death Rate', ascending=False)
```

QUESTION #3 -> Calculate the rate of people who are completely vaccinated in all counties in AL, and sort these counties by the complete vaccinated rate in descending order.

```python
df3 = pd.read_csv('COVID-19 County Vaccine Data-2.csv')
```

```python
df3.head(20)
```

```python
df3.isnull().sum()
```

```python
df3['Rate Completely Vaccinated'] = ((df3['People Completely Vaccinated']/df3['Population > 16'])*100).astype(float).round(2)
df3['% People receiving at least one dose'] = df3['% People receiving at least one dose'].astype(float).round(2)
df3['Rate No Vaccines Administered'] = (((df3['Population > 16'] - (df3['People Completely Vaccinated'] + df3['People receiving at least one dose']))/df3['Population > 16'])*100).astype(float).round(2)
df3
```

```python
df3.sort_values('Rate Completely Vaccinated', ascending=False)
```

```python
###In this barplot we can visualize the Rate Completely Vaccinated per County.

plt.figure(figsize=(15,8), dpi=200)
sns.barplot(data=df3, x='County', y='Rate Completely Vaccinated',color='orange', order=df3.sort_values('Rate Completely Vaccinated', ascending=False).County)
plt.xticks(rotation=90)
plt.ylim(0,40)
plt.show()
```

```python
###In this barplot we can visualize the Rate Completely Vaccinated per County.

plt.figure(figsize=(15,8), dpi=200)
sns.barplot(data=df3, x='County', y='Rate No Vaccines Administered',color='red', order=df3.sort_values('Rate No Vaccines Administered', ascending=False).County)
plt.xticks(rotation=90)
plt.ylim(0,100)
plt.show()
```
